```python
def readgmailmsg():
    time.sleep(10)
    # Gmail IMAP server and port
    imap_server = 'imap.gmail.com'
    port = 993
    Username = username(email= "GMail")
    Password = password(email= "GMail")
    # Your Gmail credentials
    mail = imaplib.IMAP4_SSL(imap_server, port)
    mail.login(Username, Password)

    mail.select('inbox')

    search_criteria = 'BODY "**Regards_EmailBot**"'
    # Search for all emails in the inbox
    status, data = mail.search(None, search_criteria)

    email_ids = data[0].split()

    for email_id in email_ids:
        status, msg_data = mail.fetch(email_id, "(RFC822)")
        msg_data = tuple(msg_data)[0]
        
        for response_part in msg_data:
            #if isinstance(response_part, tuple):
                # Extract the bytes object containing the email's metadata
            if isinstance(response_part, bytes):
                # Parse the email
                msg = email.message_from_bytes(response_part)
                date = msg.get("Date")
                try:
                    # Get the email subject
                    subject = decode_header(msg['Subject'])[0][0]
                    if isinstance(subject, bytes):
                        subject = subject.decode()
                    sender = decode_header(msg['From'])[0][0]
                    if isinstance(sender, bytes):
                        # Decode bytes to string
                        sender = sender.decode()
                    print('--------------------------------------')
                    print('Subject:', subject)
                    print('From:', sender)
                    print('Date:', date)
                    print('--------------------------------------')
                    
                    # Fetch the email body
                    if msg.is_multipart():
                        for part in msg.walk():
                            content_type = part.get_content_type()  # This will get the message type "text/plain if no any attachment is contained"
                            content_disposition = part.get("Content-Disposition")
                            try:
                                # Get the email body
                                body = part.get_payload(decode=True).decode()
                                #if content_type == "text/plain" and "attachment" not in content_disposition:
                                print(body)
                            except Exception:
                                pass
                    else:
                        # For non-multipart emails, get the email body
                        body = msg.get_payload(decode=True).decode()
                        print(body)
                        print('--------------------------------------')
                    deletemsgpermanentlygmail(emailid=email_id)
                except Exception:
                    pass
    
    mail.logout()
            
```
